```python
# Import libraries
import numpy as np; # Mostly used for array manipulation datatype numpy.ndarray
from sklearn.linear_model import LinearRegression # Used for implementing Logistic regression
%matplotlib notebook
import matplotlib.pyplot as plt # Plotting in python3 using Matplotlib library
import math # math module for exp function; For e value
# generate random integer values
from numpy.random import seed
from numpy.random import randint
import array
# seed random number generator
seed(1)
```

```python
def layer_analysis(inputs_activations,weights,b):
    # linear regression
    lin_reg = np.dot(inputs_activations,weights)+b 
    # Rectified Linear Unit (ReLU)
    relu_val = lin_reg
    #relu_val = max(0,lin_reg) if lin_reg >= 0
    #relu_val = 0 if lin_reg < 0
    relu_val[relu_val<0] = 0
     
    
    return relu_val
```

```python
def main():
       
    # Input feature 1 patient age
    # Input feature 2 tumor size
    # Input feature 3 family history (y/n:0/1)
    # Input feature 4 regular junk food intake
    # Input feature 5 currently smoking
    # Input feature 6 stressful lifestyle
    # Input feature 7 working with chemicals
    # Input feature 8 urban/rural (0/1)
    # Input feature 9 sleep duration (1hr-10hrs)
    # Input feature 10 regular exercise
    
    pat_age = randint(0,100,100)
    tum_size = randint(1,10,100) # in cm
    fm_histry = randint(0,2,100)
    junk_food = randint(0,2,100)
    smoking = randint(0,2,100)
    stress_lif = randint(0,2,100)
    work_withchem = randint(0,2,100)
    rural_urbn = randint(0,2,100)
    sleep_dur = randint(1,10,100)
    reg_exerc = randint(0,2,100)
    features = np.transpose([pat_age, tum_size, fm_histry, junk_food,smoking,stress_lif,work_withchem,rural_urbn,sleep_dur,reg_exerc]);
    [rows, cols] = np.shape(features)
    
    # Output benign or malignant 0/1
    diagnosis_res = randint(0,2,100)
    
    
        
    print("Enter number of layers:")
    num_oflayers = int(input())
    
    nodes_eachlayer = np.empty([num_oflayers],dtype=int)
    
    
    for eachlayer in range(num_oflayers):
       
        if(eachlayer == 0):
            nodes_eachlayer[eachlayer] = cols
        else:
            print("Enter num of nodes for layer no:", eachlayer)
            nodes_eachlayer[eachlayer] = int(input())
    #print(nodes_eachlayer)
        
    # weights for given layer B will have the dimensions: (noofnodesinlayerA X noofnodesinlayerB)
    
    for eachlayer in range(num_oflayers):
        print("For layer no:",eachlayer)
        if(eachlayer == 0): # no procesing done just setting up input nodes = number of features
            nodes_eachlayer[eachlayer] = cols
            shape = (rows,cols)
            inputs_activations = features # then input activations will be features itself shape:rowsxcols
        else:
            shape = (nodes_eachlayer[eachlayer-1],nodes_eachlayer[eachlayer])
            dt = np.float64
            weights = np.empty(shape,dtype=dt)
            val = 2
            weights.fill(val)
            #print("Dim of weights:",weights.shape)
            b = np.empty(nodes_eachlayer[eachlayer],dtype = dt)
            val_b = randint(100,300,nodes_eachlayer[eachlayer])
            # Randomize y-intercept/bias for each layer num of b vals = num of nodes in that layer 
            b = val_b
            
            ########### Start to implement logistic as a function of linear ############
            inputs_activations = layer_analysis(inputs_activations,weights,b)
            
            print(f'For layer: {eachlayer}')
            print(f'dim of activation_vector: {inputs_activations.shape}')
            print(inputs_activations)

if __name__ == "__main__":
    main()  
```
